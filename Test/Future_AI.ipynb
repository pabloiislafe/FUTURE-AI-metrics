```python
from Metrics.Robustness.dataset_variations import *

import numpy as np

y_true = np.array([0, 1, 1, 0, 1])
y_pred = np.array([0, 1, 0, 0, 1])

print("Accuracy:", compute_accuracy(y_true, y_pred))
print("Precision:", compute_precision(y_true, y_pred))
print("Recall:", compute_recall(y_true, y_pred))
print("F1:", compute_f1(y_true, y_pred))
```

```python
y_true = np.array([0, 1, 2, 2])
y_pred = np.array([0, 2, 2, 1])

print("Accuracy:", compute_accuracy(y_true, y_pred))
print("Precision (macro):", compute_precision(y_true, y_pred, average="macro"))
print("Recall (macro):", compute_recall(y_true, y_pred, average="macro"))
print("F1 (macro):", compute_f1(y_true, y_pred, average="macro"))
```

```python
mask_true = np.array([
    [1, 0, 1],
    [0, 1, 1],
    [1, 0, 0]
])

mask_pred = np.array([
    [1, 1, 1],
    [0, 1, 0],
    [1, 0, 0]
])

print("Accuracy:", compute_accuracy(mask_true, mask_pred))
print("Precision:", compute_precision(mask_true, mask_pred))
print("Recall:", compute_recall(mask_true, mask_pred))
print("F1 (Dice):", compute_f1(mask_true, mask_pred))
print("IoU:", compute_iou(mask_true, mask_pred))
print("Dice:", compute_dice(mask_true, mask_pred))
```

```python
#To evaluate only one part

ignore_mask = np.array([
    [0, 1, 0],
    [0, 0, 0],
    [1, 0, 0]
], dtype=bool)

print("Accuracy:", compute_accuracy(mask_true, mask_pred, ignore_mask=ignore_mask))
print("Precision:", compute_precision(mask_true, mask_pred, ignore_mask=ignore_mask))
print("Recall:", compute_recall(mask_true, mask_pred, ignore_mask=ignore_mask))
print("Dice:", compute_dice(mask_true, mask_pred, ignore_mask=ignore_mask))
```

```python
y_true = np.array([0, 1, 1, 0])
y_score = np.array([0.1, 0.8, 0.7, 0.3])  # model probabilities

print("ROC AUC:", compute_roc_auc(y_true, y_score))
```

```python
y_true = np.array([0, 1, 2, 1])
y_score = np.array([
    [0.7, 0.2, 0.1],  # probs for sample 1
    [0.1, 0.6, 0.3],
    [0.05, 0.25, 0.7],
    [0.2, 0.7, 0.1]
])

print("ROC AUC (macro):", compute_roc_auc(y_true, y_score, multi_class="ovr"))
```
